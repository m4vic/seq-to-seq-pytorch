```python
# tokenize
def tokenize(text):
  tect = text = text.lower()
  text.replace('?','')
  text.replace("'","")
  return text.split()
```

```python
# vocab
vocab = {'<UNK>':0}
```

```python
def build_vocab(row):
  print(row['question'])
  tokenized_question = tokenize(row['question'])
  tokenized_answer = tokenize(row['answer'])

  mearged = tokenized_question + tokenized_answer # mearging the tokenized question and as

  for word in mearged:
    if word not in vocab: # checking the word in vocab
      vocab[word] = len(vocab) # word will be added into vocab if not , as index no.


```

```python
df.apply(build_vocab,axis=1)
```

```python
# convert words to numerical indies

def word_to_index(text,vocab):

  indexed_text = []

  for token in tokenize(text): # iterate each word

    if token in vocab: # if token word in vocab word appen the indexed.text with vocab token

      indexed_text.append(vocab[token])

    else: # else append index_text with Unk index un known index

      indexed_text.append(vocab['<UNK>'])

  return indexed_text
```

```python
word_to_index('this is sanskar', vocab)  # 0 = unknown means the this and sanskar is not present in the vocab and is index is 2
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
```

```python
class QADataset(Dataset):
  def __init__(self,df,vocab):
    self.df = df
    self.vocab = vocab

  def __len__(self):
    return len(self.df)

  def __getitem__(self,idx):
   # retriving the question from dataset
    neumerical_question = word_to_index(self.df.iloc[idx]['question'],self.vocab)
    neumerical_ans = word_to_index(self.df.iloc[idx]['answer'],self.vocab)

    return torch.tensor(neumerical_question), torch.tensor(neumerical_ans)


```

```python
dataset = QADataset(df,vocab)
```

```python
dataset[2]
```

```python
dataloader = DataLoader(dataset,batch_size=1)
```

```python
for question, answer in dataloader:
  print(question, answer)

```

# *Building RNN*

```python
import torch.nn as nn
```

```python
class SimpleRNN(nn.Module):

  def __init__(self, vocab_size):
    super().__init__()
    self.embedding = nn.Embedding(vocab_size, embedding_dim=50)
    self.rnn = nn.RNN(50, 64, batch_first=True)
    self.fc = nn.Linear(64,vocab_size)

  def forward(self, question):
    embedded_question = self.embedding(question)
    hidden, final = self.rnn(embedded_question)
    output = self.fc(final.squeeze(0))

    return output
```

```python
learning_rate = 0.001
epochs = 100
model = SimpleRNN(len(vocab))
```

```python

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
```

```python
# training loop

for epoch in range(epochs):

  total_loss = 0

  for question, answer in dataloader:

    optimizer.zero_grad()

    # forward pass
    output = model(question)

    # loss -> output shape (1,337) - (1)
    loss = criterion(output, answer[0])

    # gradients
    loss.backward()

    # update
    optimizer.step()

    total_loss = total_loss + loss.item()

  print(f"Epoch: {epoch+1}, Loss: {total_loss:4f}")
```

```python
def predict(model, question, threshold=0.5):
  # convert question to neumbers
  neumerical_question = word_to_index(question, vocab)

  # tensor and send to model

  output = model(torch.tensor(neumerical_question)).unsqueeze(0)

  # convert logits to probability

  probs = torch.nn.functional.softmax(output,dim=1)

  # find the index of max prob

  value, index = torch.max(probs, dim=1)

  if value < threshold:
    print("i dont know")


  print(list(vocab.keys())[index])
```

```python
print(predict(model, "What is the capital of Japan?"))
```
