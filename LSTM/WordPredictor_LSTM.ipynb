```python
def text_to_indices(sentence, vocab): # this function will take the sentence and vocab
  numerical_sentence = []

  for token in sentence: # token in sentence
    if token in vocab: # if token is present in the vocab
      numerical_sentence.append(vocab[token]) # numerica_sentcen will be appended by the vocabs token  and vocab tokn has indices in it
    else:
      numerical_sentence.append(vocab['<unk>']) # if not than appended with <unk>

  return numerical_sentence
```

```python
input_numerical_sentences = []
for sentence in input_sentences: # here each sentence in imput sentence will be itrate and appended with indices
  input_numerical_sentences.append(text_to_indices(word_tokenize(sentence.lower()), vocab))
```

# *Dataset Dataloader object*

```python
class CustomDataset(Dataset):
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def __len__(self):
    return self.x.shape[0] # returns length

  def __getitem__(self, index): # index of x and y
    return self.x[index], self.y[index]
```

```python
dataset = CustomDataset(x,y)
```

```python
dataset[0]
```

```python
dataloader = DataLoader(dataset, batch_size=32, shuffle=True) # creating dataloader objects
```

# **LSTM-Model**

```python
class LSTMModel(nn.Module):
  def __init__(self,vocab_size):
    super().__init__()
    self.embedding = nn.Embedding(vocab_size, 100)
    self.lstm = nn.LSTM(100, 150, batch_first=True,) # 100 unputs 150 outputs
    self.ln = nn.LayerNorm(150) # layer norm
    self.fc = nn.Linear(150, vocab_size) # 128 input

  def forward(self, x): # forward propagation
    embedded = self.embedding(x) # sending to embeding layer

    intermediate_hidden_state, (final_hidden_state, final_cell_state) = self.lstm(embedded)
    # layer norm to last hidden state
    final_hidden_state = self.ln(final_hidden_state[-1]) # layer norm
    output = self.fc(final_hidden_state) # linear layer
    return output


```

```python
model = LSTMModel(len(vocab))
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
model.to(device)
```

```python
epochs = 200
learning_rate = 0.001
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
criterion = nn.CrossEntropyLoss()
```

```python
#training_loop

for epoch in range(epochs):
  total_loss = 0

  for batch_x , batch_y in dataloader:
    batch_x = batch_x.to(device) # moving to gpu
    batch_y = batch_y.to(device)

    optimizer.zero_grad() # gradient clearning
    output = model(batch_x) # output
    loss = criterion(output, batch_y) # loss calculation

    loss.backward() # back propagation
    optimizer.step() # update gradients
    total_loss = total_loss + loss.item() # loss calculation

  print('epoch', epoch, 'loss', total_loss)
```
